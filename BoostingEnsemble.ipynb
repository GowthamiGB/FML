```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.ensemble import AdaBoostClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
dataset=pd.read_csv("drive/My Drive/FML/mushrooms.csv")
data=dataset
data
data.head(5)
```

```python
dataset = data.sample(frac = 1)
dataset.colums = ['class', 'cap-shape',	'cap-surface',	'cap-color',	'bruises',	'odor',	'gill-attachment',	'gill-spacing',	'gill-size',	'gill-color']
dataset
```

```python
# Label Encoding
label_encoders = {}
for label in dataset.columns:
    label_encoders[label] = LabelEncoder()
    dataset[label] = label_encoders[label].fit_transform(dataset[label])
```

```python
#split dataset in features and target variable
X = dataset.drop(['class'],axis = 1)
y = dataset['class']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y, test_size=0.3 , random_state = 1)
```

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth=1)
AdaBoost = AdaBoostClassifier(base_estimator=model, n_estimators=400, learning_rate=1)
boostmodel = AdaBoost.fit(X_train, y_train)
```

```python
y_pred = boostmodel.predict(X_test)
```

```python
prediction = metrics.accuracy_score(y_test, y_pred)
```

```python
print("Accuracy:",prediction*100, '%')
```
